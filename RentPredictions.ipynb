```python
import pandas as pd

house_data = pd.read_csv('house_data.csv')

house_data = house_data[house_data['price'] < 12000]

house_data.dropna(axis = 0, inplace = True)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xs = house_data["arrondissement"], ys = house_data["price"], zs = house_data["surface"],  c="r", marker="o")

plt.show()

plt.plot(house_data['arrondissement'], house_data['price'], 'ro', markersize= 4)

plt.show()

plt.plot(house_data['arrondissement'], house_data['surface'], 'ro', markersize= 4)

plt.show()

plt.plot(house_data['surface'], house_data['price'], 'ro', markersize= 4)

plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = house_data.drop(columns=['price'])
y = house_data["price"]

print(X.shape)
print(y.shape)

xtrain, xtest, ytrain, ytest = train_test_split(X, y, train_size=0.8)

print("train set shape: ", xtrain.shape, ytrain.shape)
print("testing set shape: ", xtest.shape, ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression

regr = LinearRegression()

regr.fit(xtrain, ytrain)
train_score = regr.score(xtrain, ytrain)
test_score = regr.score(xtest, ytest)

print("train score: ", train_score)
print("test score: ", test_score)
```
